# Regression Interview Questions and Answers

This notebook contains commonly asked interview questions and answers on:
- Simple Linear Regression
- Multiple Linear Regression
- Polynomial Regression


## Q1. What is Simple Linear Regression?

**Answer:**

Simple Linear Regression is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.

## Q2. What are the key assumptions of Simple Linear Regression?

**Answer:**

The assumptions are linearity, independence, homoscedasticity, normality of residuals, and no significant outliers.

## Q3. What does the coefficient m represent in the equation y=mx+c?

**Answer:**

The coefficient m represents the slope of the line, showing how much y changes when x increases by 1 unit.

## Q4. What does the intercept c represent in the equation y=mx+c?

**Answer:**

The intercept c represents the value of y when x = 0.

## Q5. How do we calculate the slope m in Simple Linear Regression?

**Answer:**

The slope is calculated using covariance(X,Y) divided by variance(X).

## Q6. What is the purpose of the least squares method in Simple Linear Regression?

**Answer:**

It minimizes the sum of squared residuals between actual and predicted values.

## Q7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

**Answer:**

R² represents the proportion of variance in the dependent variable explained by the independent variable.

## Q8. What is Multiple Linear Regression?

**Answer:**

Multiple Linear Regression predicts a dependent variable using two or more independent variables.

## Q9. What is the main difference between Simple and Multiple Linear Regression?

**Answer:**

Simple regression uses one independent variable, while multiple regression uses multiple independent variables.

## Q10. What are the key assumptions of Multiple Linear Regression?

**Answer:**

Assumptions include linearity, independence, homoscedasticity, normality, and no multicollinearity.

## Q11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

**Answer:**

Heteroscedasticity means unequal variance of residuals, which can make coefficient estimates unreliable.

## Q12. How can you improve a Multiple Linear Regression model with high multicollinearity?

**Answer:**

You can remove correlated features, use PCA, or apply regularization methods like Ridge Regression.

## Q13. What are some common techniques for transforming categorical variables for use in regression models?

**Answer:**

Common methods include one-hot encoding and label encoding.

## Q14. What is the role of interaction terms in Multiple Linear Regression?

**Answer:**

Interaction terms capture combined effects of variables on the target variable.

## Q15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

**Answer:**

In multiple regression, the intercept represents the target value when all independent variables are zero.

## Q16. What is the significance of the slope in regression analysis, and how does it affect predictions?

**Answer:**

The slope indicates the direction and strength of the relationship between variables.

## Q17. How does the intercept in a regression model provide context for the relationship between variables?

**Answer:**

It provides the baseline prediction value when predictors are zero.

## Q18. What are the limitations of using R² as a sole measure of model performance?

**Answer:**

R² does not detect overfitting and may increase even with irrelevant variables.

## Q19. How would you interpret a large standard error for a regression coefficient?

**Answer:**

A large standard error indicates the coefficient estimate is less reliable.

## Q20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

**Answer:**

It appears as changing spread in residual plots and can affect statistical inference.

## Q21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

**Answer:**

It may indicate unnecessary variables are increasing R² without improving the model.

## Q22. Why is it important to scale variables in Multiple Linear Regression?

**Answer:**

Scaling helps algorithms converge faster and improves performance for some regression techniques.

## Q23. What is polynomial regression?

**Answer:**

Polynomial regression models nonlinear relationships using polynomial terms of predictors.

## Q24. How does polynomial regression differ from linear regression?

**Answer:**

Polynomial regression captures curved relationships, while linear regression models straight-line relationships.

## Q25. When is polynomial regression used?

**Answer:**

It is used when data shows a nonlinear trend.

## Q26. What is the general equation for polynomial regression?

**Answer:**

The equation includes higher-degree terms such as y = b0 + b1x + b2x² + ... + bnxⁿ.

## Q27. Can polynomial regression be applied to multiple variables?

**Answer:**

Yes, polynomial features can be created for multiple variables.

## Q28. What are the limitations of polynomial regression?

**Answer:**

It can overfit data and become computationally expensive for high degrees.

## Q29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

**Answer:**

Cross-validation, RMSE, MAE, and adjusted R² are commonly used.

## Q30. Why is visualization important in polynomial regression?

**Answer:**

Visualization helps understand nonlinear patterns and model fit.

## Q31. How is polynomial regression implemented in Python?

**Answer:**

It is commonly implemented using PolynomialFeatures from scikit-learn with LinearRegression.
